```python
from ultralytics import YOLO 
import matplotlib.pyplot as plt
import cv2
import os
%matplotlib inline

path = os.getcwd()
print(path)

im = os.path.join(path, 'Construction Site Safety.v30-raw-images_latestversion.yolov11/detect/images/test')
la = os.path.join(path, 'Construction Site Safety.v30-raw-images_latestversion.yolov11/detect/labels/test')

# Carica il modello addestrato
model = YOLO('weights/hub/11s/best.pt')

# Cartelle delle immagini e annotazioni (ground truth)
image_dir = im
label_dir = la

# Lista delle immagini da elaborare
image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))]

plot_data = []

# Itera sulle immagini del test set
for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    label_path = os.path.join(label_dir, image_file.replace('.jpg', '.txt').replace('.png', '.txt'))

    # Carica l'immagine
    orig_img = cv2.imread(image_path)
    
    # Se ci sono annotazioni, le carica
    im_gt = orig_img.copy()
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                class_id, x_center, y_center, width, height = map(float, line.strip().split())
                img_h, img_w, _ = orig_img.shape
                
                # Converti da YOLO format a coordinate (x1, y1, x2, y2)
                x1 = int((x_center - width / 2) * img_w)
                y1 = int((y_center - height / 2) * img_h)
                x2 = int((x_center + width / 2) * img_w)
                y2 = int((y_center + height / 2) * img_h)

                # Disegna la bounding box ground truth (rosso)
                cv2.rectangle(im_gt, (x1, y1), (x2, y2), (255, 0, 0), 2)  

    # Esegui la predizione del modello
    results = model(image_path)

    # Immagine con bounding box predette
    im_pred = orig_img.copy()
    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Coordinate bounding box predetta
            conf = box.conf[0].item()  # Confidence score
            
            # Disegna la bounding box
            cv2.rectangle(im_pred, (x1, y1), (x2, y2), (0, 255, 0), 2)  
            
            # Aggiungi il testo con la confidence score
            label = f'{conf:.2f}'
            cv2.putText(im_pred, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Converti immagini in RGB per matplotlib
    orig_img = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
    im_gt = cv2.cvtColor(im_gt, cv2.COLOR_BGR2RGB)
    im_pred = cv2.cvtColor(im_pred, cv2.COLOR_BGR2RGB)

    plot_data.append((orig_img, im_gt, im_pred))

# Crea la figura e gli assi
fig, axes = plt.subplots(len(plot_data), 3, figsize=(15, 5 * len(plot_data)))

# Itera sui dati e crea i plot
for i, (orig_img, gt_img, pred_img) in enumerate(plot_data):
    axes[i, 0].imshow(orig_img)  # Immagine originale
    axes[i, 0].set_title('Immagine originale')
    axes[i, 0].axis('off')

    axes[i, 1].imshow(gt_img)  # Ground truth
    axes[i, 1].set_title('Bounding Box Reali (GT)')
    axes[i, 1].axis('off')

    axes[i, 2].imshow(pred_img)  # Bounding box predette
    axes[i, 2].set_title('Bounding Box Predette')
    axes[i, 2].axis('off')

# Mostra il plot
plt.tight_layout()
plt.show()
```
